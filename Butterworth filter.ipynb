```python
from scipy.signal import *
```

Butterworth default

```python
normalised_passband_freq = 2  # equal to 1 for standard Butterworth
```

```python
import matplotlib.pyplot as plt
plt.figure(1)
plt.title('Magnitude plot of Butterworth filters of increasing order')
for N in range(1,10):
    (num,denom)=butter(N,normalised_passband_freq,'lowpass' ,True)
    sys = TransferFunction(num,denom)
    w, mag, phase = sys.bode()
    #plt.figure()
    plt.semilogx(w, mag,label=N)    # Bode magnitude plot
    plt.legend()
    plt.grid()
    #plt.figure()
    #plt.semilogx(w, phase)  # Bode phase plot

plt.figure(2)
#plt.legend()
plt.title('Phase of Butterworth fiters of increasing order')
for N in range(1,10):
    (num,denom)=butter(N,normalised_passband_freq,'lowpass' ,True)
    sys = TransferFunction(num,denom)
    w, mag, phase = sys.bode()
    #plt.figure()
    #plt.semilogx(w, mag)    # Bode magnitude plot
    #plt.figure()
    plt.semilogx(w, phase,label=N)  # Bode phase plot
    plt.legend()
    
plt.show()
```
