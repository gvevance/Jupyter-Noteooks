```python
from scipy.signal import *
```

Chebyshev filter

```python
normalised_passband_freq = 2 #equal to 1 for standard
passband_ripple = 1 #in dB the allowed passband ripple
```

```python
import matplotlib.pyplot as plt
plt.figure(1)
plt.title('Magnitude plot of Chebyshev Type I filters of increasing order')
for N in range(1,5):
    (num,denom)=cheby1(N,passband_ripple,normalised_passband_freq,'lowpass' ,True)
    sys = TransferFunction(num,denom)
    w, mag, phase = sys.bode()
    #plt.figure()
    plt.semilogx(w, mag,label=N)    # Bode magnitude plot
    plt.legend()
    #plt.xlim(0.01,1.2)
    #plt.ylim(-2,1)
    #plt.figure()
    #plt.semilogx(w, phase)  # Bode phase plot

plt.figure(2)
#plt.legend()
plt.title('Phase of Chebyshev Type I fiters of increasing order')
for N in range(1,5):
    (num,denom)=cheby1(N,passband_ripple,normalised_passband_freq,'lowpass' ,True)
    sys = TransferFunction(num,denom)
    w, mag, phase = sys.bode()
    #plt.figure()
    #plt.semilogx(w, mag)    # Bode magnitude plot
    #plt.figure()
    plt.semilogx(w, phase,label=N)# Bode phase plot
    #plt.xlim(0.01,1.2)
    #plt.ylim(-8,1)
    plt.legend()
    
plt.show()
```
