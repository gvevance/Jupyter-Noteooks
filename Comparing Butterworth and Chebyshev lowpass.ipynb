```python
from scipy.signal import *
```

```python
normalised_passband_freq = 1  # equal to 1 for standard Butterworth
passband_ripple = 3 #in dB the allowed passband ripple
```

```python
import matplotlib.pyplot as plt
```

```python
N=int(input("Enter order of filter : "))
```

```python
(num,denom)=butter(N,normalised_passband_freq,'lowpass' ,True)
sys = TransferFunction(num,denom)

(num2,denom2)=cheby1(N,passband_ripple,normalised_passband_freq,'lowpass' ,True)
sys2 = TransferFunction(num2,denom2)


#Plotting magnitudes
plt.figure(1)
plt.title('Magnitude')

w, mag, phase = sys.bode()
w, mag2, phase2 = sys2.bode()
plt.semilogx(w, mag,label='%dth order Butterworth'%N)    # Bode magnitude plot
plt.semilogx(w, mag2,label='%dth order Chebyshev'%N)    # Bode magnitude plot
plt.legend()
plt.xlim(0.01,1.2)

#Uncomment to see ripples in Chebyshev
plt.ylim(-4,1)

#Plotting phases
plt.figure(2)
plt.title('Phase')
plt.semilogx(w,phase,label='%dth order Butterworth'%N)
plt.semilogx(w,phase2,label='%dth order Chebyshev'%N)
plt.legend()
plt.xlim(0.01,1.2)


```
