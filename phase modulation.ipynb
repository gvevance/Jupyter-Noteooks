```python
import numpy as np
import matplotlib.pylab as plt
import scipy.signal as sig
plt.rcParams['figure.figsize'] = [20, 10]
```

```python
t=np.linspace(0,10,1000)
```

```python
kp=1
Ac=2
fc=1000
Am=1
```

```python
# m=Am*np.cos(2*np.pi*100*t)
```

```python
m=sig.square(2*np.pi*200*t,0.33333)
```

```python
theta=kp*m
```

```python
c = Ac*np.cos(2*np.pi*fc*t)
```

```python
PM_wave=Ac*np.cos(2*np.pi*fc*t + theta)
```

```python
plt.subplot(221)
plt.plot(t,PM_wave,'g')
plt.xlabel('t')
plt.ylabel("PM carrier wave and carrier wave")

plt.subplot(223)
plt.xlabel('t')
plt.ylabel("carrier wave")
plt.plot(t,c,'r')

plt.subplot(222)
plt.xlabel('t')
plt.ylabel("message wave")
plt.plot(t,m,'r')
```
